```python
grouped_by_year = df_2006_2022.groupby('Jahr')

# Initialize an empty dictionary to store the DataFrames for each year
teil_datensets = {}

# Iterate through each group (based on the year) and store the DataFrame in the dictionary
for year in range(2006, 2023):  # Including 2006 and up to 2022
    teil_datensets[year] = grouped_by_year.get_group(year)

# Now you have separate DataFrames for each year from 2006 to 2022
# Accessing a specific DataFrame for a particular year, e.g., for the year 2006:
teil_datenset_2006 = teil_datensets[2006]
teil_datenset_2007 = teil_datensets[2007]
teil_datenset_2008 = teil_datensets[2008]
teil_datenset_2009 = teil_datensets[2009]
teil_datenset_2010 = teil_datensets[2010]
teil_datenset_2011 = teil_datensets[2011]
teil_datenset_2012 = teil_datensets[2012]
teil_datenset_2013 = teil_datensets[2013]
teil_datenset_2014 = teil_datensets[2014]
teil_datenset_2015 = teil_datensets[2015]
teil_datenset_2016 = teil_datensets[2016]
teil_datenset_2017 = teil_datensets[2017]
teil_datenset_2018 = teil_datensets[2018]
teil_datenset_2019 = teil_datensets[2019]
teil_datenset_2020 = teil_datensets[2020]
teil_datenset_2021 = teil_datensets[2021]
teil_datenset_2022 = teil_datensets[2022]

for year in range(2006, 2023):
    teil_datensets[year]['Dichte'] = (teil_datensets[year]['Basiswert 1'] / teil_datensets[year]['Basiswert 2']) 

teil_datenset_2006.head(40)
```

```python
import geopandas as gpd


# Read the GeoJSON file
gdf = gpd.read_file('output_postal.geojson')

```

```python
import json

# Iterate through each year
for year, teil_datenset in teil_datensets.items():
    # Open the GeoJSON file
    with open('output_postal.geojson', 'r') as f:
        geojson_data = json.load(f)
    
    # Iterate through each feature in the GeoJSON file
    for feature in geojson_data['features']:
        # Get the district name for the current feature
        district_name = feature['properties']['Name']
        
        # Debugging: Print the district name
        print("District name:", district_name)
        
        # Initialize the density value as a default value
        density_value = -1  # You can choose any default value here
        
        # Check if the district name exists in the DataFrame
        if district_name in teil_datenset['Raumbezug'].values:
            # Get the corresponding density value from the DataFrame
            density_value = teil_datenset.loc[teil_datenset['Raumbezug'] == district_name, 'Dichte'].iloc[0]
        
        # Update the 'density' attribute in the 'properties' section
        feature['properties']['density'] = density_value
    
    # Save the updated GeoJSON data to a new file
    with open(f'output_postal_{year}_with_density.geojson', 'w') as f:
        json.dump(geojson_data, f)
```
